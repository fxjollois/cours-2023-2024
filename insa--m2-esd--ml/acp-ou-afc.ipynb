```python
import pandas
import numpy
import matplotlib.pyplot as plt
import seaborn
seaborn.set_style("white")

from prince import PCA
from prince import CA
```

```python
iris = pandas.read_table("https://fxjollois.github.io/donnees/Iris.txt", sep = "\t")
```

```python
iris2 = iris.drop(columns = "Species")
```

```python
pca = PCA()
pca.fit(iris2)
```

```python
pca.plot_row_coordinates(iris2, color_labels = iris["Species"])
```

```python
ca = CA()
ca.fit(iris2)
```

```python
ca_df = ca.row_coordinates(iris2)
ca_df["Species"] = iris["Species"]
ca_df
```

```python
seaborn.scatterplot(x = 0, y = 1, data = ca_df, hue = "Species")
```
